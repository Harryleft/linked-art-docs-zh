
## personDiedAtPlace

### SPARQL


```
SELECT DISTINCT ?person WHERE {
   BIND(<%current%>as ?where)
   ?person crm:P100i_died_in ?death .
    ?death crm:P7_took_place_at ?where .
 }
```
